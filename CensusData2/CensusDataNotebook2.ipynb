### Import Modules ###

```python
import pandas as pd
import numpy as np
```

### Import CSV Files and Organize Dataframes ###

```python
counties_state_list = pd.read_csv("Counties.csv")

counties_state_list.head()
```

```python
counties_state_select = counties_state_list[["County", "Metro_Area", "State"]]

counties_state_select.head()
```

```python
population_county = pd.read_csv("PopulationCounty.csv")

population_county.head()
```

```python
population_select = population_county[["County", "State", "Population"]]

population_select.head()
```

```python
income_county = pd.read_csv("IncomeCounty.csv")

income_county.head()
```

```python
income_select = income_county[["County", "State", "Personal Income(2017)", "Personal Income(Rank in State 2017)"]]

income_select.head()
```

### Merge Dataframes ###

```python
#Dataframes:
#counties_state_select
#population_select
#income_select

cspopulation = counties_state_select.merge(population_select, left_on=["County", "State"], right_on=["County", "State"], how="left")

cspopulation.head()
```

```python
cspopulation_income = cspopulation.merge(income_select, left_on=["County", "State"], right_on=["County", "State"], how="left")

cspopulation_income.head()
```

```python
cspopulation_income.to_csv("CountyStatePopulationIncome.csv")
```
